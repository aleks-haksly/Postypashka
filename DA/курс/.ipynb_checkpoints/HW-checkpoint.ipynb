## Задание 3. Модель - 30 баллов

В данном задании вам предстоит построить классифицирующую модель размер алмаза и его стоимости, а затем оценить полученную модель.

>**Примечание**
>
>Данная модель не претендует на самый верный индикатор стоимости алмаза, ведь мы не учитываем ничего больше чем стоимость алмаза. Исключительно для образовательных целей :)

Для построения модели нам понадобится 4 столбца, а именно:
- `meas_length`
- `meas_width`
- `meas_depth`
- `total_sales_price`

От остальных столбцов можно избавиться.

(1 балл) Создайте новый датасет и выведите его первые 10 строк:

```python
# your code here
data = pd.read_csv("diamonds.csv")[
    ["meas_length", "meas_width", "meas_depth", "total_sales_price"]
]

data.iloc[:10]
```

(1 балл) Перемешайте строки в датафрейме случайным образом:

```python
# your code here
data_shuffled = data.sample(frac=1, random_state=42).reset_index(drop=True)
data_shuffled.head(5)
```

(1 балл) Сформируем новую колонку - для каждого бриллианта найдем сумму Минковского (p=3) по 3 измерениям бриллианта:

```python
# your code here
data_shuffled["minkowski_sum"] = (
    abs(data_shuffled.meas_length - data_shuffled.meas_width) ** 3
    + abs(data_shuffled.meas_length - data_shuffled.meas_depth) ** 3
    + abs(data_shuffled.meas_width - data_shuffled.meas_depth) ** 3
) ** (1 / 3)
data_shuffled
```

(1 балл) Сформируем новый датасет из двух колонок - сумма Минковского и `total_sales_price`:

```python
# your code here
data_merge = data_shuffled[["minkowski_sum", "total_sales_price"]]
```

(1 балл) Разделите вашу выборку на две части в соотношении 8:2 - одна выборка будет для тренировка, а другая для тестов вашей модели:

```python
# your code here
train_data = data_merge.iloc[:int(len(data)*0.8)]
test_data = data_merge.iloc[int(len(data)*0.8):]
```

(2 балла) Постройте график зависимости стоимости от вашего нового параметра размера для тренинг-выборки:

```python
# your code here
with sns.axes_style('whitegrid'):
    plt.figure(figsize=(8, 5))
    ax = sns.scatterplot(data=train_data,y='total_sales_price', x='minkowski_sum')
    ax.set (xlabel='Minkowski',
    ylabel='Price',
    title='Price.(Minkowski)')
```

(2 балла) Нанесите на тот же график тестовую выборку, используя другой цвет (не забудьте про параметр прозрачности):

```python
# your code here
with sns.axes_style('whitegrid'):
    plt.figure(figsize=(8, 5))
    ax = sns.scatterplot(data=train_data,y='total_sales_price', x='minkowski_sum', color='red')
    sns.scatterplot(data=test_data,y='total_sales_price', x='minkowski_sum', color='blue')
    ax.set(xlabel='Minkowski',
    ylabel='Price',
    title='Price.(Minkowski)')
    plt.legend(title='Data', loc='upper right', labels=['Train Data', 'Test Data'])
```

(3 балла) Найдите для каждой бриллианта из тестовой выборки его "ближайшего соседа" из тренинг-выборки. Для каждого бриллианта из тестовой выборки найдите расстояния до соответствующего ближайшего соседа:

```python
# your code here
merged_data = pd.merge_asof(
    test_data.sort_values("minkowski_sum"),
    train_data.sort_values("minkowski_sum"),
    on="minkowski_sum",
    direction="nearest",
)
merged_data["dist"] = (
    merged_data.total_sales_price_x - merged_data.total_sales_price_y
).abs()
```

(3 балла) Найдите насколько сильно (в процентах) отличается цена бриллиантов из тестовой выборки от их соответствующих ближайших соседей:

```python
# your code here
merged_data['percent'] = (
    (merged_data.total_sales_price_x / merged_data.total_sales_price_y) * 100 - 100
)
merged_data
```

**Данная модель является выдумкой автора. Не важен коэффициент ошибки, важны ваши шаги для получения коэффициента.**

(3 балла) Мы работаем с категориальными признаками, для них не понятно что есть "больше" и "меньше". Поэтому нам нужно предварительно закодировать значения наших признаков, с чем нам поможет [OneHotEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html):

```python
# your code here
from sklearn.tree import DecisionTreeRegressor
import sklearn
import sklearn.metrics
import sklearn.model_selection
import sklearn.preprocessing
import sklearn.tree
#categ = ['symmetry', 'clarity']
#categ = ['color', 'clarity']
categ = ['cut_quality', 'clarity']
data = pd.read_csv("diamonds.csv")[[*categ, "total_sales_price"]].dropna()
for col in data.columns:
    data = data[data[col] != "unknown"]

coder = sklearn.preprocessing.OneHotEncoder(handle_unknown="ignore")
coded = coder.fit_transform(data[[*categ]])
coder.categories_
coder.get_feature_names_out([*categ])
prices = data.total_sales_price
data = pd.DataFrame(
    coded.toarray(), columns=coder.get_feature_names_out(), index=data.index
)
data
```

(3 балла) Разбейте датасет на обучающую и тестовую выбору. Для этого вам поможет [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html):

```python
# your code here
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    data, prices, test_size=0.1
)
```

(3 балла) Затем создайте и обучите вашу модель - [DecisionTreeRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html):

```python

clf = DecisionTreeRegressor(max_depth=20, min_samples_leaf=10)
clf.fit(x_train, y_train)
predicted = clf.predict(x_test)
```

(3 балла) Оцените ошибку вашей модели при помощи [mean_squared_error](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html):

```python
# your code here
print(sklearn.metrics.mean_squared_error(y_test, predicted))
```
